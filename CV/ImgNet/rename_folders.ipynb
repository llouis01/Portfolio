## Rename training folders to their human names ##

```python
import os
import csv
from pathlib import Path
```

```python
label_data = "C:/Users/RoiMinuit/Desktop/data/ILSVRC/Labels/imagenet_class_index.csv"
train_path = Path("C:/Users/RoiMinuit/Desktop/data/ILSVRC/Data/CLS-LOC/train")
```

```python
with open(label_data, 'r') as f:
    reader = csv.reader(f)
    next(reader)

    for row in reader:
        folder_name, description = row
        print(folder_name, description)
        old_folder = train_path / folder_name
        new_folder = train_path / description

        if not new_folder.exists():
            os.rename(old_folder, new_folder)
        else:
            print(f"Folder '{new_folder}' already exists.")
```
